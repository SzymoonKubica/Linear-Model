```python
map = {names:i for i,names in enumerate(films['genre'].unique())}

films['genre'] = films['genre'].replace(map)
```

```python
films['country']= films['country'].str.split(',').str[0]
```

```python
mapCountry = {names:i for i, names in enumerate(films['country'].unique())}
films['country'] = films['country'].replace(mapCountry)
```

```python
unique_names =films['directors'].unique()
name_to_value = {name: i for i, name in enumerate(unique_names)}

films['directors'] = films['directors'].map(name_to_value)
```

```python
def PolecajkaFilmowaDirichlet(titl:str):
    filmRow = films[films['title'] == titl ]
    if len(filmRow) == 0:
        return "Movie not found"
    movies = films.drop(filmRow.index)

    filmYear = filmRow['year'].values
    filmCountry = filmRow['country'].values
    filmDirector = filmRow['directors'].values
    filmGenre = filmRow['genre'].values

    filmDist = movies['year'].values - filmYear
    countryDist = movies['country'].values - filmCountry
    directorDist = movies['directors'].values - filmDirector
    genreDist = movies['genre'].values - filmGenre
    
    moviesRecomend = movies

    #Dirichlet's metric 

    moviesRecomend['year'] = moviesRecomend['year'].apply(lambda x: 1 if x != 0 else 0) 
    moviesRecomend['country'] = moviesRecomend['country'].apply(lambda x: 1 if x != 0 else 0)
    moviesRecomend['directors'] = moviesRecomend['directors'].apply(lambda x: 1 if x != 0 else 0)
    moviesRecomend['genre'] = moviesRecomend['genre'].apply(lambda x: 1 if x != 0 else 0)

    moviesRecomend['summary'] = moviesRecomend['year'] + moviesRecomend['country'] + moviesRecomend['directors'] + moviesRecomend['genre']

    bestRecomendation = moviesRecomend['summary'].sort_values(ascending=True)

    bestFilmIndex = bestRecomendation.index[0]

    bestFilm = moviesRecomend['title'][0]

    return bestFilm
```
